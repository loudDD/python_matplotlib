```python
#coding=utf-8
import matplotlib.pyplot as plt
import random
plt.figure(figsize=(20,30))
plt.plot([x for x in range(60)],[random.randint(13,25) for i in range(60)],linestyle="-.",label='')
plt.yticks(range(0,36,5))
plt.xticks(range(0,60,5),["11{}".format(i) for i in range(0,60,5)])
plt.legend()
plt.grid(True,linestyle='--',alpha=0.5)
plt.xlabel("")
plt.ylabel("")
plt.title("11")
plt.plot([x for x in range(60)],[random.randint(1,3) for i in range(60)],linestyle=':')
plt.savefig("test.png")
plt.show()
```

```python
import numpy as np
x = np.linspace(-100,100,10)
y = 2*x*x

import matplotlib.pyplot as plt
plt.figure()
plt.bar(x,y,width=10,align="center")
plt.grid(True,linestyle="--",alpha=0.5)
plt.show()
```
